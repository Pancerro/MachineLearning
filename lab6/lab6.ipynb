```
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
reader = open("/content/drive/My Drive/AdrianLawecki.txt",'r')
try:
  print(reader.name)
  type(reader)
finally:
  reader.close()
with open("/content/drive/My Drive/AdrianLawecki.txt", 'rb') as reader:
  print(reader.read())
with open("/content/drive/My Drive/AdrianLawecki.txt", 'rb') as reader:
  print(reader.readline(1))
  print(reader.readline(1))
  print(reader.readline(1))
f = open("/content/drive/My Drive/AdrianLawecki.txt")
f.readlines()
f.close
file = open("/content/drive/My Drive/AdrianLawecki.txt")
list(file)
file.close

with open('/content/drive/My Drive/AdrianLawecki.txt', 'r') as reader:
  line = reader.readline()
  while line != '':
    print(line, end='')
    line = reader.readline()

with open('/content/drive/My Drive/AdrianLawecki.txt', 'r') as reader:
  for line in reader.readlines():
    print(line, end='')

with open('/content/drive/My Drive/AdrianLawecki.txt', 'r') as reader:
  for line in reader:
    print(line, end='')

with open('/content/drive/My Drive/AdrianLawecki.txt', 'r') as reader:
    adrian = reader.readlines()

with open('/content/drive/My Drive/AdrianLawecki.txt', 'w') as writer:
    for line in reversed(adrian):
        writer.write(line+"1")


with open('/content/drive/My Drive/AdrianLawecki.png', 'rb') as byte_reader:
  print(byte_reader.read(1))
  print(byte_reader.read(3))
  print(byte_reader.read(2))
  print(byte_reader.read(1))
  print(byte_reader.read(1))


with open('/content/drive/My Drive/AdrianLawecki.txt', 'a') as a_writer:
    a_writer.write('\nAdrian Lawecki')

with open('/content/drive/My Drive/AdrianLawecki.txt', 'r') as reader:
  print(reader.read())

d_path = '/content/drive/My Drive/AdrianLawecki.txt'
d_r_path = '/content/drive/My Drive/AdrianLawecki2.txt'
with open(d_path, 'r') as reader, open(d_r_path, 'w') as writer:
    ad = reader.readlines()
    writer.writelines(reversed(ad))


class my_file_reader():
    def __init__(self, file_path):
        self.__path = file_path
        self.__file_object = None

    def __enter__(self):
        self.__file_object = open(self.__path)
        return self

    def __exit__(self, type, val, tb):
        self.__file_object.close()

with my_file_reader('/content/drive/My Drive/AdrianLawecki.txt') as reader:
  pass


class PngReader():
    _expected_magic = b'\x89PNG\r\n\x1a\n'

    def __init__(self, file_path):
        if not file_path.endswith('.png'):
            raise NameError("File must be a '.png' extension")
        self.__path = file_path
        self.__file_object = None

    def __enter__(self):
        self.__file_object = open(self.__path, 'rb')

        magic = self.__file_object.read(8)
        if magic != self._expected_magic:
            raise TypeError("The File is not a properly formatted .png file!")

        return self

    def __exit__(self, type, val, tb):
        self.__file_object.close()

    def __iter__(self):
        return self

    def __next__(self):
        initial_data = self.__file_object.read(4)
        if self.__file_object is None or initial_data == b'':
            raise StopIteration
        else:
            chunk_len = int.from_bytes(initial_data, byteorder='big')
            chunk_type = self.__file_object.read(4)
            chunk_data = self.__file_object.read(chunk_len)
            chunk_crc = self.__file_object.read(4)
            return chunk_len, chunk_type, chunk_data, chunk_crc



with PngReader('/content/drive/My Drive/AdrianLawecki.png') as reader:
  for l, t, d, c in reader:
    print("PNG")
    print(f"{l:05}, {t}, {c}")

import argparse
import os

def str2unix(input_str: str) -> str:
  r_str = input_str.replace('\r\n', '\n')
  return r_str


def dos2unix(source_file: str, dest_file: str):
  with open(source_file, 'r') as reader:
    dos_content = reader.read()
  unix_content = str2unix(dos_content)
  with open(dest_file, 'w') as writer:
    writer.write(unix_content)
if __name__ == "__main__":
  parser = argparse.ArgumentParser(
    description="Script that converts a DOS like file to an Unix like file",
    )
  parser.add_argument(
    'source_file',
    help='The location of the source '
    )
  parser.add_argument(
    '--dest_file',
    help='Location of dest file (default: source_file appended with `_unix`',
    default=None
    )
  args = parser.parse_args()
  s_file = args.source_file
  d_file = args.dest_file
  if d_file is None:
    file_path, file_extension = os.path.splitext(s_file)
    d_file = f'{file_path}_unix{file_extension}'
  dos2unix(s_file, d_file)
```
